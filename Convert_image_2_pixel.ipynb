```python
# Python script to convert image to C array
from PIL import Image
import numpy as np

img = Image.open('happy_overlay.png').convert('RGB') # Replace with any image you like! 
img = img.resize((128, 128)) 
pixels = np.array(img)

# Generate C code
print("const uint32_t satisfied_data[128][128] = {")
for row in pixels:
    print("  {", end="")
    for r, g, b in row:
        rgb = (b << 16) | (g << 8) | r  # BGR format
        print(f"0x{rgb:06X}, ", end="")
    print("},")
print("};")
```
